```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
df = pd.read_csv("C:\\Users\\sridhar\\Desktop\\stock market forecasting\\invest.csv")
```

```python
df
```

```python
X = df[['Monthly Income', 'Monthly Expenditure']]
y = df['Invest']
```

```python
X
```

```python
y
```

```python
logistic_regression= LogisticRegression()
logistic_regression.fit(X,y)
```

```python
pickle.dump(logistic_regression, open('Finalmodel.pkl','wb'))
```
